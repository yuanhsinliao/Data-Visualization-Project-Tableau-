```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Dataset /german_credit.csv")
```

```python
df
```

```python
df.isnull().any()
```

```python
df = pd.get_dummies(df, columns=["Payment Status of Previous Credit", "Purpose", "Most valuable available asset", "Occupation", "Telephone", "Foreign Worker", "Sex and Marital Status",'Account Balance'])
```

```python
sns.countplot(df,x="Creditability")
```

```python
sns.catplot(df,x="Creditability",y="Credit Amount", kind="violin")
```

```python
X = df.drop(columns="Creditability")
Y = df["Creditability"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y)
```

```python
from scipy import stats
X_train["Credit Amount"] = stats.zscore(X_train["Credit Amount"])
X_test["Credit Amount"] = stats.zscore(X_test["Credit Amount"])
```

```python
from sklearn import linear_model
```

```python
from pandas.core.common import random_state
model = linear_model.LogisticRegression(max_iter=2000)
```

```python
model.fit(X_train,Y_train)
pred = model.predict(X_test)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, pred)
print(cm)

print("Accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```

```python
import matplotlib.pyplot as plt

from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_predictions(Y_test, pred)
plt.show()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
pred = model.predict(X_test)
cm = confusion_matrix(Y_test, pred)
print(cm)
print("Accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```

```python
from sklearn import ensemble
```

```python
model = ensemble.RandomForestClassifier()
```

```python
model.fit(X_train,Y_train)
pred = model.predict(X_test)
cm = confusion_matrix(Y_test, pred)
print(cm)
print("Accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```

```python
fr = model.feature_importances_
print(fr)

for importance, name in sorted(zip(model.feature_importances_,X_train.columns),reverse=True)[:5]:
    print (name, importance)
```

```python
model = ensemble.GradientBoostingClassifier()
```

```python
model.fit(X_train,Y_train)
pred = model.predict(X_test)
cm = confusion_matrix(Y_test, pred)
print(cm)
print("Accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```

```python
from sklearn import neural_network
```

```python
model = neural_network.MLPClassifier()
```

```python
model.fit(X_train,Y_train)
pred = model.predict(X_test)
cm = confusion_matrix(Y_test, pred)
print(cm)
print("Accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```

```python
from lce import LCEClassifier
# Train LCEClassifier with default parameters
model = LCEClassifier()
model.fit(X_train, Y_train)
pred = model.predict(X_test)
cm = confusion_matrix(Y_test, pred)
print(cm)
print("Accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout

model = Sequential()
model.add(Dense(100,input_dim=47,activation="relu"))
model.add(Dropout(0.1))
model.add(Dense(100,activation="relu"))
model.add(Dropout(0.1))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
h = model.fit(X_train,Y_train,batch_size=10,epochs=300)

import matplotlib.pyplot as plt
plt.plot(h.history["loss"])

pred = model.predict(X_test)
import numpy
pred = numpy.where(pred>=0.5,1,0)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,pred)
print(cm)
print("accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```
